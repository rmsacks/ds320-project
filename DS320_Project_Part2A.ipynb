```
import datascience
from datascience import *

import pandas as pd
import numpy as np
import spacy
import re

path1= "/content/drive/My Drive/president_county.csv"
pres_county = Table.read_table(path1)
path2 = "/content/drive/My Drive/president_county_candidate.csv"
pres_county_cand = Table.read_table(path2)

#Converted tables to data frames
pres_county = pres_county.to_df()
pres_county_cand = pres_county_cand.to_df()

#Got rid of missing values with .dropna()
pres_county_full = pres_county.replace(r'^\s*$', np.nan, regex=True).dropna(axis=0)
pres_county_cand_full = pres_county_cand.replace(r'^\s*$', np.nan, regex=True).dropna(axis=0)

#Got rid of dups with .drop_duplicates()
pres_county_full = pres_county_full.drop_duplicates(subset=None, keep='first', inplace=False)
pres_county_cand_full = pres_county_cand_full.drop_duplicates(subset=None, keep='first', inplace=False)
```

```
#Running the first Jaccard similarity measure on the 'state' column
col_vals = set(pres_county_full['state'].unique())
for col in pres_county_cand_full.columns:
    ext_col_vals = set(pres_county_cand_full[col].unique())
    intersection_size = len(col_vals.intersection(ext_col_vals))
    union_size = len(col_vals.union(ext_col_vals))
    jaccard = intersection_size / union_size
    if jaccard != 0:
        print(col + ' has similarity score ' + str(jaccard))
```
